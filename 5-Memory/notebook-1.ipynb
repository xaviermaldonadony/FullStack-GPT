```python
from langchain.memory import ConversationBufferWindowMemory 

memory = ConversationBufferWindowMemory(
    return_message = True,
    k = 4
)

def add_message(input, output):
    memory.save_context({"input":input},{"outputs": output})

add_message(1,1)
```

```python
add_message(2,2)
add_message(3,3)
add_message(4,4)
```
